### Introduction

Today, a company's website is akin to a virtual storefront, providing customers with crucial information about its offerings. However, with the ever-increasing amount of data on these websites, analyzing thousands of companies belonging to a particular industry can be overwhelming. This makes it challenging for customers to gain a comprehensive understanding of the industry and each individual company.

Text analytics is a powerful tool that can offer valuable insights into analyzing the vast amount of data on company websites. Text analytics can help in various ways, from identifying similarities between companies to detecting whether a company exists in the real world or is generated by an algorithm like Chat GPT.

In this report, we focus on identifying the general characteristics of company websites and comparing them across companies that produce magnets under the same NAICS code. By doing so, we aim to gain a deeper understanding of the industry, its trends, and opportunities for growth.

### PCA

Principal Component Analysis (PCA) is a useful technique for gaining an understanding of the underlying structure of company websites. By analyzing the word distributions of each principal component, we can see that these components are primarily composed of two types of words: website configuration-related words and product-related words.

Specifically, when examining PC0, we can observe that the `pos` column is predominantly composed of website configuration-related words (cookies, domain, website, forge, information, privacy), whereas the `neg` column is mostly comprised of product-related words (die, casting, aluminum, zinc, castings, sand).

<img src= "code/PCA_0.png">

The loadings plot is a valuable tool for visualizing the two primary axes of company websites. The right side of the plot is dominated by website configuration-related words, such as cookies, domain, data, information, and policy. In contrast, the left side of the plot is primarily associated with product-related words, including casting, aluminum, mold, and sand.

<img src = "code/PCA_1.png">

The plot below provides an interesting insight into the relationship between the number of links on a company's website (size of a dot) and the content of that website. Specifically, the size of each dot on the plot, which is projected onto PC0 and PC1, is proportional to the number of links on the website. As the value of PC0 increases, so does the size of the dot, indicating that companies with more links are more likely to have website-configuration-related content.

Most companies in the magnets manufacturing industry tend to focus on their products and capabilities, which occupy a relatively small portion of their websites. Instead, the majority of their websites contain non-product-related information such as contact pages, board of directors, privacy notices, and protocols. This trend is clearly visible in the plot.

However, there are some exceptions to this trend, as evidenced by the smaller dots on the far-right side of the plot. Although companies with more links tend to have more non-product-related information on their websites, this is not always the case. There may be several reasons for this, such as different configuration settings or privacy protocols that hinder automatic web scraping. Therefore, there is significant variance in the number and type of links that are scraped for different companies.

<img src = "code/PCA_2.png">

### Word Embedding

Word Embedding is an effective method for capturing the relationship between words within a corpus. Given that we already know from PCA that the corpus is primarily divided into two axes, website configurations and products, I used Word Embedding to investigate whether the model can differentiate between these two distinct axes.

To build the Word Embedding model, I used the following hyperparameters: 
- `window` = 5
- `vector_size` = 50
- `min_count` = 20
- `workers` = 4 

Since company websites typically cover a limited range of topics, I reduced the `vector_size` to avoid overfitting to the semantic meaning of each word.

The t-SNE plot below confirms the results from PCA, showing two distinct clusters. The top left region of the plot contains words related to manufacturing, such as aluminum, casting, steel, and brass. In contrast, the bottom right region is dominated by words related to website configurations, such as information, privacy, and data.

<img src = 'code/EMBEDDING_1.png'>

To identify the hierarchy of words, clustering plots were added to the analysis. The plots revealed two major clusters, which are consistent with the two distinct axes identified in the previous analysis. When looking at individiual words, this plot successfully captures highly related words: alloy - alloys, machine - tooling, copper - bronze, contact - phone, and address - email.

<img src = 'code/EMBEDDING_2.png'>

### LDA

Latent Dirichlet Allocation (LDA) allows us detect the topics that are embedded within our corpus, using DOCUMENT-TOPIC MATRIX (`THETA`) and TERM-TOPIC MATRIX (`PHI`). Since we already know that there exist only small number of topics within company websites, I extracted only 10 topics. The hyperparameters I used are:

- `ngram_range` = [1,2]
- `n_terms` = 4000
- `n_topics` = 10
- `max_iter` = 20
- `n_top_terms` = 9

According to the table below, the topic keywords are familar to us, and we can cleary detect the two main topic axes here again: products and websites. Interestingly, we can identify another topic (`T05`). This topic is mainmly about business: quality, customers, products, company, years, industry, customer, buisness, and manufacturing. 

<img src = "code/LDA_0.png">

It is surprising that the business topic (`T05`) was not identified during PCA and Word Embedding, yet based on the term frequency, it was the most dominant topic in our corpus according to LDA.

<img src = "code/LDA_2.png">

Additionally, based on the doc weighted sum, the business topic (`T05`) was also the most dominant topic in our corpus. 

<img src = "code/LDA_1.png">

From this plot, we can identify that there is a positive relationship between term frequency and document weight sum. This explains the general similarity ranks between the two.

<img src = "code/LDA_3.png">

Overall, these results show that LDA is a powerful technique for identifying topics within a corpus and can reveal insights that other techniques may miss.

### Sentiment Analysis

Company websites are typically written in a formal and straightforward language, making it difficult to discern any emotional fluctuations within the text. To gauge the extent of positive emotions such as trust, anticipation, and joy present within company websites, I conducted sentiment analysis.

To assess the general distribution of emotion within the corpus, I calculated the mean emotion score for all companies. As expected, the dominant emotions were trust and anticipation. However, it was surprising to find that sadness and fear were also well represented in the corpus.

<img src="code/SENTIMENT_1.png">

In this dataframe, we can observe the assigned emotions for each term. It is noteworthy that the word "die" has been classified as negative, despite its contextual meaning within our corpus as a specialized tool for shaping or cutting materials. Similarly, the word "iron" has been classified as positive, contrary to our expectations. On the other hand, the word "hood", which can refer to a protective covering, has been classified as negative in our dataset, but it typically has a neutral connotation.

<img src = "code/SENTIMENT_0.png">

To visualize the words associated with each emotion, I decided to use a word cloud generator rather than the conventional visualization techniques.

In this plot, I generated a word cloud for `company_id = 858`, which was marked as having the highest level of anger. However, upon inspection, it was difficult to identify any words directly related to anger. The only possible candidate was the word "cut," which is an ambiguous word that can be interpreted as a threat to a person. In the context of the company's website, however, it likely refers to the process of cutting materials to produce magnets.

<img src = 'code/WORDCLOUD_ANGER.jpeg'>

In this plot, I generated a word cloud for `company_id = 284`, which was marked as having the highest level of anticipation. However, upon inspection, it was difficult to identify any words directly related to anticipation. Most of the words with a large font size were related to the company name and were irrelevant to our analysis, such as artworks, studio, and art.

<img src = "code/WORDCLOUD_ANTICIAPTION.png">

In this plot, I generated a word cloud for `company_id = 5`, which was marked as having the highest level of trust. While some of the words in the word cloud, such as information, contact, service, tolerances, manufacturing, and quality, were quite related to trust, they did not appear frequently enough to strongly indicate trust. Surprisingly, some of the words with the largest font size, such as noblitt, blog, fabrication, and fabricating, were not directly related to trust.

<img src = "code/WORDCLOUD_TRUST.png">

The sentiment analysis model's inability to accurately detect emotions can be attributed to examples such as the plot shown below. In our specific context, the word "die" does not have its common meaning of "death." Instead, it typically refers to a specialized tool used to shape or cut materials into specific shapes for company websites producing magnets. However, our model fails to detect the contextual meaning of the word "die," leading to a critical misclassification of emotions that blurs our interpretation of the data. As shown in the second dataframe plot, the word 'die' occurred the most, a total of 563 times, which can significantly distort the sentiment analysis of our corpus.

## Conclusion

This project was able to explore the general content of company websites manufacturing magnets.

In terms of PCA, we had two main axes comprising company websites: products and website configuration. Also, the more links a website had, the more it is likely to contain website configuration related content. On the other hand, the lesser links a website had, the more likely it is likely to have product related content.

The result of word embedding did align with PCA. The t-SNE plot clearly displays the disconnection between two topics, on the top left, there existed a cluster mostly comprised of production related words. On the bottom right, there was a cluster for website related words. 

Surprisingly, LDA captured the hidden topic of our corpus: business. It provides an extra point of view, other than the known topics, proudcts and webistes. LDA, in fact, did really capture the latent topic space that was not identified in PCA and word embedding.

For sentiment analysis, I initially expected trust and anticipation would be the dominant emotion, which was proved by the mean emotion of all companies. However, when I dived deep into a company level using wordcloud, the word combinations were not relevant with the emotion itself, for anger, turst, and anticiapation. Also, the word 'die' was misclassified, regardless of the context of our corpus, leading to the overall misinterpreation of sentiment within the corpus.

To summarize, this project successfully identified the two main semantic axes of company websites producing magnets through PCA and word embedding, and revealed an additional topic through LDA. However, there were limitations to this project due to the lack of direct or chronological order in company websites, making it impossible to capture the sequential order of emotion and plot, and the absence of linkage between links, making it difficult to detect connections between them. Nonetheless, this textual analysis can serve as a valuable reference for individuals seeking to gain a better understanding of companies beyond standardized data such as company reports or stock trends.
